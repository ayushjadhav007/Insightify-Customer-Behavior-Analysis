```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv("customer_shopping_behavior.csv")
```

```python
df.head(5)
```

```python
df.isnull().sum()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower() # convert all Capitals with lower case
df.columns = df.columns.str.replace(' ','_') # Replace the space with underscore
```

```python
df.columns
```

```python
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#Create a column called age Groups
labels = ['Young Adult','Adult','Middle-aged','Senior']
df['age_group'] = pd.qcut(df['age'],q=4 ,labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# create new column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['Frequency of Purchases'].map(frequency_mapping)
```

```python

df[['purchase_frequency_days','Frequency of Purchases']].head(10)
```

```python
df[['Discount Applied','Promo Code Used']].head(10)
```

```python

(df['Discount Applied'] == df['Promo Code Used']).all()
```

```python
# Dropping promo code used column

df = df.drop('Promo Code Used', axis=1)
```
